Q1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

*Examples:*  
`find_the_difference("abcd", "abcde") ➞ "e"`  
`find_the_difference("", "y") ➞ "y"`  
`find_the_difference("ae", "aea") ➞ "a"`  

```python
def find_the_difference(s,t):
    t_temp = list(t)
    for ele in s:
        if ele in t_temp:
            t_temp.pop(t_temp.index(ele))
    return  ','.join(t_temp)
```

```python
print('{} -> {}'.format('find_the_difference("abcd", "abcde")',find_the_difference("abcd", "abcde")))
print('{} -> {}'.format('find_the_difference("", "y")',find_the_difference("", "y")))
print('{} -> {}'.format('find_the_difference("ae", "aea")',find_the_difference("ae", "aea")))
```

Q2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

List order is: `[int, str, bool, list, tuple, dictionary]`  

*Examples:*   
`count_datatypes(1, 45, "Hi", False) ➞ [2, 1, 1, 0, 0, 0]`  
`count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]`  
`count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2]`  
`count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) ➞ [2, 0, 1, 2, 2, 0]`  

```python
def count_datatypes(*args):
    count = {'int':0,'str':0,'bool':0,'list':0,'tuple':0,'dict':0}
    for ele in args:
        count[type(ele).__name__] += 1
    return list(count.values())
```

```python
print('{} -> {}'.format('count_datatypes(1, 45, "Hi", False) ',count_datatypes(1, 45, "Hi", False) ))
print('{} -> {}'.format('count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ',count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ))
```

Q3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.

*Examples:*   
`fib_str(3, ["j", "h"]) ➞ "j, h, hj"`  
`fib_str(5, ["e", "a"]) ➞ "e, a, ae, aea, aeaae"`  
`fib_str(6, ["n", "k"]) ➞ "n, k, kn, knk, knkkn, knkknknk"`  

```python
def fib_str(num,chars):
    prev2, prev = chars[0],chars[1]
    fibonacci_str=", ".join(chars)
    if num > 2:
        for ele in range(2,num):
            temp = prev+prev2
            fibonacci_str += ", "+temp
            prev2 = prev
            prev = temp
    return fibonacci_str
```

```python
print(f'fib_str(3, ["j", "h"]) -> "{fib_str(3, ["j", "h"])}"')
print(f'fib_str(5, ["e", "a"]) -> "{fib_str(5, ["e", "a"])}"')
print(f'fib_str(6, ["n", "k"]) -> "{fib_str(6, ["n", "k"])}"')
```

Q4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:
`"nines:your answer, threes:your answer, ones:your answer"`  
You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).  

*Examples:*    
`ones_threes_nines(10) ➞ "nines:1, threes:0, ones:1"`  
`ones_threes_nines(15) ➞ "nines:1, threes:2, ones:0"`  
`ones_threes_nines(22) ➞ "nines:2, threes:1, ones:1"`  

```python
def ones_threes_nines(num):

    answer = {'nines':0,'threes':0,'ones':0}

    while num > 0 and num < 26:
        if num >= 9:
            answer['nines'] += 1
            num -= 9
        elif num >= 3:
            answer['threes'] += 1
            num -= 3
        elif num >= 1:
            answer['ones'] +=1
            num -=1
    return answer
```

```python
print(f'ones_threes_nines(10) -> "{ones_threes_nines(10)}"')
print(f'ones_threes_nines(15) -> "{ones_threes_nines(15)}"')
print(f'ones_threes_nines(22) -> "{ones_threes_nines(22)}"')
```

Q5. The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci sequence recursively can grow quite fast. Therefore it can be important to think about the implications of running a function recursively. Depending on the size of n needed and the capabilities of the system in question you might want to take a different approach.
Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n. 

*Examples:*    
`fib(6) ➞ 8`  
`# 0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8`  
`fib(1) ➞ 1`  
`fib(2) ➞ 1`  

```python
def fib(in_num):
    prev2, prev = 0, 1
    for ele in range(2,in_num+1):
        temp = prev+prev2
        prev2 = prev
        prev = temp
    return prev
    
```

```python
print(f'fib(6) -> {fib(6)}')
print(f'fib(1) -> {fib(1)}')
print(f'fib(2) -> {fib(2)}')
```
